```python
import numpy as np
import pandas as pd
```

```python
data = np.genfromtxt('iris.data.txt', delimiter=',', usecols=(0,1,2,3))
```

1. Exiba (print) a primeira coluna (atributo) de todas as amostras

```python
print(data[:,0])
```

2. Faça o mesmo que acima considerando somente o tipo Versicolor

```python
print(data[50:100, 0])
```

3. Considerando o que já estudamos, faça um gráfico como o abaixo


```python
## Comprimento da Sépala IRIS_SETOSA 
## Comprimento da Sépala IRIS_VERSICOLOR

from matplotlib import pyplot as plt
setosa = data[:50,0]
versicolor = data[50:100,0]
plt.plot([i for i, _ in enumerate(setosa)], setosa, 'b-', label='Comprimento da Sépala IRIS_SETOSA')
plt.plot([i for i, _ in enumerate(versicolor)], versicolor, 'r-.', label='Comprimento da Sépala IRIS_VERSICOLOR')
plt.legend(loc=1)
plt.title("Comprimento da Sépala")
plt.show()
```

4. Reconstrua o gráfico acima considerando as espécies Setosa e Virgínica

```python
## Comprimento da Sépala IRIS_SETOSA 
## Comprimento da Sépala IRIS_Virgínica

from matplotlib import pyplot as plt
setosa = data[:50,0]
virginica = data[100:,0]
plt.plot([i for i, _ in enumerate(setosa)], setosa, 'b-', label='Comprimento da Sépala IRIS_SETOSA')
plt.plot([i for i, _ in enumerate(virginica)], virginica, 'r-.', label='Comprimento da Sépala IRIS_Virgínica')
plt.legend(loc=1)
plt.title("Comprimento da Sépala")
plt.show()
```

5. Faça um novo gráfico, semelhante ao acima, considerando os atributos comprimento da
pétala entre as espécies Setosa e Virgínica.

```python
## Comprimento da pétala IRIS_SETOSA 
## Comprimento da pétala IRIS_Virgínica

from matplotlib import pyplot as plt
setosa = data[:50,2]
virginica = data[100:,2]
plt.plot([i for i, _ in enumerate(setosa)], setosa, 'b-', label='Comprimento da pétala IRIS_SETOSA')
plt.plot([i for i, _ in enumerate(virginica)], virginica, 'r-.', label='Comprimento da pétala IRIS_Virgínica')
plt.legend(loc=1)
plt.title("Comprimento da Pétala")
plt.show()
```

6. Repita o gráfico acima para as espécies Versicolor e Virginica.

```python
## Comprimento da pétala IRIS_Versicolor
## Comprimento da pétala IRIS_Virgínica

from matplotlib import pyplot as plt
versicolor = data[50:100,2]
virginica = data[100:,2]
plt.plot([i for i, _ in enumerate(versicolor)], versicolor, 'b-', label='Comprimento da pétala IRIS_VERSICOLOR')
plt.plot([i for i, _ in enumerate(virginica)], virginica, 'r-.', label='Comprimento da pétala IRIS_Virgínica')
plt.legend(loc=1)
plt.title("Comprimento da Pétala")
plt.show()
```

Comparacao comprimento de todas as cepalas

```python
## Comprimento da Sépala IRIS_SETOSA 
## Comprimento da Sépala IRIS_Versicolor
## Comprimento da Sépala IRIS_Virgínica

from matplotlib import pyplot as plt
setosa = data[:49,0]
versicolor = data[50:99,0]
virginica = data[100:,0]
plt.plot([i for i, _ in enumerate(setosa)], setosa, 'b-', label='Comprimento da Sépala IRIS_SETOSA')
plt.plot([i for i, _ in enumerate(versicolor)], versicolor, 'g-', label='Comprimento da Sépala IRIS_VERSICOLOR')
plt.plot([i for i, _ in enumerate(virginica)], virginica, 'r-.', label='Comprimento da Sépala IRIS_Virgínica')
plt.legend(loc=1)
plt.title("Comprimento da Sépala")
plt.show()
```

imprima um grafico de dispersao comparando o comprimento e a largura da petala de todas as especies


```python
from matplotlib import pyplot as plt
comprimento = data[:,2]
largura = data[:,3]

labels = ['setosa', 'versicolor', 'virginica']
plt.scatter(comprimento, largura)

plt.title("Comprimento x largura das petalas")
plt.xlabel("Comprimento")
plt.ylabel("Largura")
plt.axis("equal")
plt.show()
```
